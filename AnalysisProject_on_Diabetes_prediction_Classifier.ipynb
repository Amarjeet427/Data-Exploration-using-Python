# 8.Diabetic Cases Pie Chart
This pie chart shows how and why the spread of diabetic classes is important. It gives a short visual overview that helps you understand how the information is put together, which is important for doing good research and making good decisions in data visualization projects.

```python
import matplotlib.pyplot as plt

# Assuming a binary class label named "Diabetic"
target_variable = "class"

# Count occurrences of each class label
class_counts = df[target_variable].value_counts()

# Create a pie chart to visualize the class distribution
plt.figure(figsize=(8, 6))
plt.pie(class_counts, labels=class_counts.index, autopct="%1.1f%%", startangle=140)
plt.title("Distribution of Diabetic Cases", fontsize=16)

plt.show()
```

# 10.Aging vs. Health Metrics: Box Plots
Box plots for "Tri Fold Thick," "Serum Ins," "BMI," and "DP Function" by age make it easier to find outliers. This visualization is very important for finding health trends linked to getting older, planning treatments, and making healthcare better in projects that use data visualization.








```python
for y_var in ["Tri Fold Thick", "Serum Ins", "BMI", "DP Function"]:
    sns.boxplot(x="Age", y=y_var, showmeans=True, data=df)
    plt.title(f"Distribution of {y_var} by Age (Outlier Detection)", fontsize=16)
    plt.xlabel("Age", fontsize=12)
    plt.ylabel("Value", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# 12.PG Concentration Impact on Health Parameters
When you compare health indicators like "Pregnancies," "Diastolic BP," and "Tri Fold Thick" by PG concentration, you can see how it affects health. This information helps us understand how glucose levels affect different parts of health, which is important for customizing treatments and getting better results in healthcare projects that use data visualization.







```python
feature_to_compare = "PG Concentration"
threshold = 140
high_glucose_data = df[df[feature_to_compare] >= threshold]
low_glucose_data = df[df[feature_to_compare] < threshold]
plt.figure(figsize=(16, 12))
for i, feature in enumerate(["Pregnancies", "Diastolic BP", "Tri Fold Thick"], start=1):
    plt.subplot(3, 3, i)
    sns.kdeplot(high_glucose_data[feature], label="High PG Concentration", color="red", shade=True)
    sns.kdeplot(low_glucose_data[feature], label="Low PG Concentration", color="blue", shade=True)
    plt.title(f"Distribution of {feature} by PG Concentration", fontsize=16)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel("Density", fontsize=12)
    plt.legend()
plt.tight_layout()
plt.show()
```

# 14.Age-BMI Relationship: Regression Plot
A joint plot with a regression line that shows the link between age and BMI. This image shows how age and BMI are related, which helps us understand health trends and plan specific actions to improve healthcare results in data visualization projects.








```python
x_var = "Age"
y_var = "BMI"
sns.jointplot(x=x_var, y=y_var, data=df, kind="reg")
plt.title(f"Relationship between {x_var} and {y_var} (Regression Line)", fontsize=16)
plt.xlabel(x_var, fontsize=12)
plt.ylabel(y_var, fontsize=12)
plt.show()
```
